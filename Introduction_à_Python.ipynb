## **3.** Conversion de type

Les variables en Python ne sont pas typées. C'est l'affetation d'une valeur à une variable qui attribue implicitement un type à une variable : `int`, `float`, `str`, booléen, liste.

Si un calcul fait intervenir un réel ou donne un résultat réel (division, racine carrée, etc.), alors Python convertit implicitement les entiers en rèels avant le calcul.

Il est parfois nécessaire de forcer la conversion d'une variable.

```
n = 2             # affectation d'un entier (int)
r = 10.0          # affectation d'un réel (float)
c = "Bonjour"     # affectation d'une chaine de caractères (string)
b = True          # affectation d'un booléen

d = r/n           # Python convertit implicitement l'entier `n` en réel (float)
print(d)          # 5.0

ch1 = "32"
ch2 = "16"
ch3 = ch1 + ch2   # concaténation de chaines de caractères
print(ch3)        # '3216'

k = int(ch1) + int(ch2)   # conversion en entier (int)
print(k)                  # 48

print(b)                  # True
```

Il est aussi possible de convertir un nombre entier (**integer**) ou réel (**float**) en chaine de caractères (**string**).

```
n = 10                # n est est une variable de type entier
ch1 = str(n)          # ch1 est une variable de type string
pi = 3.14             # pi est de type float
ch2 = str(pi)         # ch2 est de type string
print( n + pi )       # addition de nombres réels : 13.14
print( ch1 + ch2 )    # concaténation de chaines de caractères : 103.14
print( ch1 + pi )     # Erreur ! string + float
```

## **6.** Booléens et expressions conditionnelles

Un booléen est un type de variable à deux états (généralement notés Vrai et Faux), destiné à représenter les valeurs de vérité de la logique booléenne. Il est nommé ainsi d'après George Boole, fondateur dans le milieu du xixe siècle de l'algèbre portant son nom et créateur de la logique moderne. Le type de données booléen est principalement associé à des états conditionnels. 

Une proposition est une phrase avec une affirmation qui peut être vraie ou fausse, aussi appelée **expression conditionnelle**.

*Exemples :*
- "2 est un chiffre pair" est une proposition vraie.
- "5 est un multiple de 2" est une proposition fausse.

*Remarque :* Une proposition ne peut jamais être à la fois vraie et fausse.

Dans une instruction conditionnelle, on établit la valeur d'une proposition (Vraie ou Fausse) pour exécuter une (ou des) instruction(s). 

On peut aussi associer la valeur d'une proposition à une variable. Cette variable obtient alors le type booléen.




```
5>3       # est une expression conditionnelle, vraie dans ce cas
```

```
7<=2
```

```
(2+3)==7      # est une autre expression conditionnelle, fausse dans ce cas
```

```
# On peut associer une expression conditionnelle (ou proposition) à une variable
# La variable prend la valeur de la proposition : Vrai, ici.
a = 5>3
print("a =",a)

# La variable 'a' est alors de type booléen
# La fonction 'type' retourne le type d'une variable : booléen, ici
print(type(a))
```

## **10.** Boucle non-bornée `while`

Une boucle non-bornée permet de répéter une suite d'instructions tant qu'une *condition* est vérifiée. Ce type de boucle est utilisé lorsque nombre de répétitions n'est pas connu à l'avance.

**Exemple :** tirer un nombre aléatoire entre 1 et 10. On arrête dès que le nombre tiré au sort est supérieur ou égal à 7.


```
import random # import de la bibliothèque de fonctions aléatoires

nombre_aleatoire=random.randint(1,10)
print(nombre_aleatoire, end=' ')    # Le paramètre end=' ' permet de supprimer le retour à la ligne après l'affichage

while nombre_aleatoire < 7 :
  nombre_aleatoire=random.randint(1,10)
  print(nombre_aleatoire, end=' ')
```

## **11.** Fonctions (blocs d'instructions) `def`

Une **fonction** est un bloc d'instructions séparé du programme principal. 
Ce bloc de code pourra être exécuté plusieurs fois dans le programme principal par un appel à la fonction.

Une fonction peut retourner une valeur au programme principal (un peu comme une fonction mathématique).

*Attention :* les variables définies dans une fonction restent internes à la fonction. Ce sont des "variables locales".
Les variables définies dans le programme principal ne sont pas visibles dans une fonction. 

Les paramètres d'une fonction permettent de transmettre des valeurs à la fonction.

**Exemple :** afficher le tableau de valeur d'une fonction affine.

```
# Cette fonction de paramètre x calcule l'image d'un antécédent x par une fonction affine
def fonction_affine(x):
  m = -2
  p = 15
  y = m*x + p
  return y      # valeur retournée par la fonction

# Programme principal
print ("x f(x)")

for a in range(1,11) :
  print(a, fonction_affine(a))    # appel à la fonction
                                  # la valeur "a" est transmise à la fonction
```

**Exemple :** tester si un nombre est pair ou impair.

```
# Cette fonction teste si un nombre donné comme paramètre est pair ou impaire
def pair_impair(n):
  if n%2 == 0:
    return "est un nombre pair !"
  else:
    return "est un nombre impair !"

# Programme principal
for i in range(3):
  nombre = int(input("Entrez un nombre entier : "))
  print(nombre, pair_impair(nombre))
```

### Détermination de la longueur d'une courbe

```
# Calcul de la longueur d'une courbe sur un intervalle [a;b]
#
from math import *
import numpy as np
import matplotlib.pyplot as plt

# Exemples de fonctions
def f(x):
  y = x*x                         # fonction carré
#  y = np.sin(x)                   # sinusoïde
#  y = np.sin(x)*np.cos(3*x)       # signal modulé
#  y = x**3*np.sin(x)/(1+2*x**2)   # oscillations amorties 
  return y

def norme(x,y):
  distance = np.sqrt(x*x + y*y)
  return distance

# bornes de l'intervalle
a = -1
b = 1
# nombre de pas
n = 4
# distance d'un pas
d = (b-a)/n
# calcul de la valeur approchée de la longeur de la courbe de la fonction f 
# sur [a;b] avec un pas de d
longueur = 0
for i in range(0,n):
  x1 = a + i*d
  x2 = x1 + d
  longueur = longueur + norme(d,f(x2)-f(x1))

print("Longueur =", longueur)

affiche_graphique = True
if affiche_graphique :
  plt.style.use('default') # Style de graphique prédéfini (défault)
  plt.rcParams["figure.figsize"] = (10,7)  # taille du graphique en inches
  
  # Tableau des valeurs de x et calcul des valeurs de la fonction
  x = np.linspace(a, b, 200)
  y = f(x)
  # Tableau des n valeurs de x et calcul des valeurs de la fonction
  xn = np.linspace(a, b, n+1)
  yn = f(xn)

  # Tracé des points (x,y) reliés par des segments
  plt.plot(x,y,label='$f(x)$')
  plt.plot(xn,yn,label='$f$ approchée par ' + str(n) + ' segments \nLongueur = ' + str(longueur))

  # Affichage de la légende
  plt.legend() 
  plt.grid(True)
  plt.xlabel('$x$')
  plt.ylabel('$f(x)$')
  plt.show()
```

### Calcul de la dimension fractale d'une courbe par la méthode box-counting

```
# ==============================================================================
# Calcule une valeur approchée de la dimension fractale (dimension d'Hausdorff) 
# d'une courbe par la méthode de Minkowski (box-counting).
# Citation : A procedure to Estimate the Fractal Dimension of Waveforms, 
# Carlos Sevcik, 1998, Complexity International
# Updated paper on arXiv:1003.5266v1 (2010)
# Sevcik-1998-A procedure to Estimate the Fractal Dimension of Waveforms.pdf
#

import numpy as np
import matplotlib.pyplot as plt

# Initialisation de la série aléatoire
# pour effectuer la même simulation
#np.random.seed(123)

# bornes de l'intervalle
a = 0
b = 1
# nombre de points de la courbe
n = 10000

x = np.linspace(a, b, n)

# Exemples de courbes
courbe = 7    # choix de la courbe

if courbe == 0 :
  y = x                            # fonction linéaire
if courbe == 1 :
  y = x*x                          # fonction carré
elif courbe == 2 :
  y = np.sin(x)                    # sinusoïde
elif courbe == 3 :
  y = np.sin(x)*np.cos(3*x)        # signal modulé
elif courbe == 4 :
  y = x**3*np.sin(x)/(1+2*x**2)    # oscillations amorties 
elif courbe == 5 :
  y = np.random.normal(size=n)     # bruit blanc gaussien
elif courbe == 6 :
  y = np.random.random(size=n)     # bruit blanc uniforme
elif courbe == 7 :                 # mouvement brownien
  # https://github.com/mCodingLLC/VideosSampleCode/blob/master/brownian_motion.py
  dt = x[1] - x[0]
  # Bt2 - Bt1 ~ Normal with mean 0 and variance t2-t1
  dy = np.sqrt(dt) * np.random.normal(size=(n-1,))
  y0 = np.zeros(shape=(1,))
  y  = np.concatenate((y0, np.cumsum(dy)))

  def quadratic_variation(y):
    return np.cumsum(np.power(np.diff(y, prepend=0.), 2))

# Détermination de ymin et ymax
ymin = y[0]
ymax = ymin

for i in range(1,n):
  if y[i] > ymax : ymax = y[i]
  if y[i] < ymin : ymin = y[i]
  
# Calcul optimisé de la longueur normalisée de la courbe
dx2   = (1.0 / (n-1))**2
ay     = ymax - ymin
y1     = y[0]
length = 0.0  
for i in range(1,n):
  y2 = y[i]
  length = length + np.sqrt( ((y2-y1)/ay)**2 + dx2 )
  y1 = y2
  
# calcul de la dimension fractale
dimf = 1.0 + (np.log(length) - np.log(2)) / np.log(2*(n - 1))

print("Dimension fractale =", dimf)

# Tracé de la courbe
plt.style.use('default') # Style de graphique prédéfini (défault)
plt.rcParams["figure.figsize"] = (12,6)  # taille du graphique en inches

plt.plot(x, y, linewidth=1)
if courbe == 7 : plt.plot(x, quadratic_variation(y))

plt.figtext(0.15, 0.9, '$D_H$ = ' + str(dimf))

plt.grid(True)
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```

### Convergence de l'estimation de la dimension fractale $D_H$

```
# Convergence de l'estimation de la dimension fractale par box-counting

import numpy as np
import matplotlib.pyplot as plt

# bornes de l'intervalle
a = 0
b = 1
# nombre de points de la courbe
ns = 2000

xn = range(10,ns+1)
dh = []

for n in xn :
  np.random.seed(123)   # Initialisation de la série aléatoire
  x = np.linspace(a, b, n)
  # Exemples de courbes
  courbe = 0    # choix de la courbe

  if courbe == 0 :
    y = x                            # fonction linéaire
  if courbe == 1 :
    y = x*x                          # fonction carré
  elif courbe == 2 :
    y = np.sin(x)                    # sinusoïde
  elif courbe == 3 :
    y = np.sin(x)*np.cos(3*x)        # signal modulé
  elif courbe == 4 :
    y = x**3*np.sin(x)/(1+2*x**2)    # oscillations amorties 
  elif courbe == 5 :
    y = np.random.normal(size=n)     # bruit blanc gaussien
  elif courbe == 6 :
    y = np.random.random(size=n)     # bruit blanc uniforme
  elif courbe == 7 :                 # mouvement brownien
    # https://github.com/mCodingLLC/VideosSampleCode/blob/master/brownian_motion.py
    dt = x[1] - x[0]
    # Bt2 - Bt1 ~ Normal with mean 0 and variance t2-t1
    dy = np.sqrt(dt) * np.random.normal(size=(n-1,))
    y0 = np.zeros(shape=(1,))
    y  = np.concatenate((y0, np.cumsum(dy)))

  # Détermination de ymin et ymax
  ymin = y[0]
  ymax = ymin

  for i in range(1,n):
    if y[i] > ymax : ymax = y[i]
    if y[i] < ymin : ymin = y[i]
    
  # Calcul optimisé de la longueur normalisée de la courbe
  dx2   = (1.0 / (n-1))**2
  ay     = ymax - ymin
  y1     = y[0]
  length = 0.0  
  for i in range(1,n):
    y2 = y[i]
    length = length + np.sqrt( ((y2-y1)/ay)**2 + dx2 )
    y1 = y2

  # calcul de la dimension fractale
  dh.append( 1.0 + (np.log(length) - np.log(2)) / np.log(2*(n - 1)) )

# Tracé de la courbe de convergence
plt.style.use('default') # Style de graphique prédéfini (défault)
plt.rcParams["figure.figsize"] = (12,6)  # taille du graphique en inches

plt.plot(xn, dh, linewidth=1)

plt.grid(True)
plt.xlabel('$n$')
plt.ylabel('$D_H$')
plt.show()
```

Déterminer la fréquence de l'événement A au bout de plusieurs essais. 
Calculer la probabilité de l'événement A.

Simulation de plusieurs essais.

```
import random

nb_essais = int(input("Nombre d'essais ? "))
effectif_6 = 0
for i in range(1,nb_essais+1):
  de1 = random.randint(1,6)
  de2 = random.randint(1,6)
  de3 = random.randint(1,6)
  de4 = random.randint(1,6)
  if de1 == 6 or de2 == 6 or de3 == 6 or de4 ==6 :
    effectif_6 += 1
  frequence_6 = effectif_6/i
  print("Fréquence   = ", frequence_6)

p = 1 - (5/6)**4
print("Probabilité = ", p)
```

## **21.** Fluctuations d'échantillonnage

### Répétition d'expérience aléatoire

On simule 100 échantillons de taille *n* associés à une expérience aléatoire dont la probabilité de succès est *p*. 

On trace le graphique donnant la fréquence *f* de succès obtenue pour chaque échantillon.

On constate que la fréquence *f* de succès varie d'un échantillon à l'autre : c'est la fluctuation d'échantillonnage.

```
import random
import numpy as np
import matplotlib.pyplot as plt

# Style de graphique prédéfini par défault
plt.style.use('default') 

def trace_points(f,n):
    plt.plot([n],[f],'b.')

p = float(input("Entrer la probabilité p à simuler: "))
n = int(input("Entrer la taille n des échantillons à simuler: "))

print("Variations obtenues : ", 1/np.sqrt(n))
    

for i in range(1,101):
    effectif = 0
    for j in range(0,n):
        if random.random() <= p :
            effectif = effectif + 1
    trace_points(effectif/n,i)

plt.axis([0,100,0,1])
plt.grid(True)
plt.xlabel('Echantillon n°')
plt.ylabel('Fréquence dans l\'échantillon')
plt.show()
```

### Fluctuation des fréquences en fonction de la taille de l'échantillon

On simule 1000 échantillons de taille 1 à 1001 associés à une expérience aléatoire dont la probabilité de succès est *p*. 

Le graphique donne la fréquence *f* de succès obtenue pour chaque échantillon en fonction de la taille de l'échantillon.
On observe que les variations de la fréquence *f* observée sont d'autant plus petites que la taille de l'échantillon est grande.

```
import random
import numpy as np
import matplotlib.pyplot as plt

# Style de graphique prédéfini par défault
plt.style.use('default')
plt.rcParams["figure.figsize"] = (18,9)

def trace_points(f,n):
    plt.plot([n],[f],'b.')

p = float(input("Entrer la probabilité p à simuler: "))
#p = random.randint(10,90)/100    
    

for i in range(1,1001):
    effectif = 0
    for j in range(0,i):
        if random.random() <= p :
            effectif = effectif + 1
    trace_points(effectif/i,i)

# Création d'une liste x de 100 valeurs
x = np.linspace(1, 1000, 100)
# Définition des enveloppes
y1 = p + 1/np.sqrt(x)
y2 = p - 1/np.sqrt(x)
# Tracé des enveloppes            
plt.plot(x,y1,'y--',label='$y_1(n)=p+\dfrac{1}{\sqrt{n}}$')   # Tracé de la courbe y1 avec légende (avec LaTeX)
plt.plot(x,y2,'g--',label='$y_2(n)=p-\dfrac{1}{\sqrt{n}}$')

# Probabilité p
plt.hlines(p, 1, 1000, colors='red', zorder=3, label="Probabilité $p$ = " +  str(p))

# Affichage de la légende
plt.legend() 
# Définition des axes
plt.axis([0,1000,0,1])
plt.grid(True)
plt.xlabel("Taille de l'échantillon")
plt.ylabel("Fréquence dans l'échantillon")
plt.show()
```

## **22.** Simulation Monte Carlo

On peut obtenir une estimation de $\pi$ en tirant des nombres aléatoires.

On tire au sort les coordonnées de points dans un carré. Le rapport du nombre de points dans le cercle inscrit sur le nombre de points dans le carré est proportionnel au rapport des surfaces, ce qui permet de calculer une valeur approchée de $\pi$.

Aire du carré de coté 2 $= 2\times 2 = 4$.

Aire du cercle de rayon 1 $= \pi R^2 = \pi$.

Rapport des surfaces $= \dfrac\pi4 = \dfrac{n_1}{n}$, où $n_1$ est le nombre de points tombés dans le cercle et $n$ le nombre de points dans le carré.

```
from math import pi
import random
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle, Circle

n = 10000  # nombre de points aléatoires

# Tracé du graphique seulement si n < 10000
if n <= 10000 : 
  plt.style.use('default')  # Style de graphique prédéfini (défault)
  plt.rcParams["figure.figsize"] = (7,7)  # taille du graphique en inches
  # plt.gca().set_aspect('equal', adjustable='box') # Ratio des axes
  plt.gca().add_patch(Rectangle((-1, -1), 2, 2, color="aliceblue")) # Carré
  plt.gca().add_patch(Circle((0, 0), radius=1, color="red", alpha=0.2)) # Disque

# Fonction de tracé du nuage de points en couleur
def trace_points(x,y,color):
    plt.plot([x],[y], color=color, marker='.', markersize=2)

k = 0  # compteur du nombre de points à l'intérieur du disque

for i in range(n):
  x = random.uniform(-1,1)
  y = random.uniform(-1,1)
  if x*x + y*y < 1 :
    k += 1
    if n <= 10000 : trace_points(x,y,"red")
  else :
    if n <= 10000 : trace_points(x,y,"darkgray")

if n <= 10000 : 
  plt.hlines(0, -1.5, 1.5, color='gray', linewidth=1) # axe des abscisses
  plt.vlines(0, -1.5, 1.5, color='gray', linewidth=1) # axe des ordonnées
  plt.axis([-1.5,1.5,-1.5,1.5])
  plt.grid(True)
  plt.show()

print("Estimation = ", k/n*4)
print("Pi         = ", pi)
```
